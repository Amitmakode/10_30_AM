#### 18.1.1.3 Histograms

it was introduced by *Karl Pearson*. It is a representation of numerical data in a form of bar graphs. Following is a bried explaination: -

* BINS: these are nothing division of entire range of values into some specific series of intervals. These intervals are non overlapping. 
* Then we plot the count of values or frequency of values which fall in these bins or buckets in form of a bar

For more information you can refer - [Histogram (WIKIPEDIA)](https://en.wikipedia.org/wiki/Histogram)

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the iris dataset
iris = datasets.load_iris()
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Create a distribution plot with a curve using seaborn
sns.set(style="whitegrid")
for column in iris_data.columns:
    sns.kdeplot(iris_data[column], shade=True)

plt.title("Distribution Plot with Curve for Iris Dataset")
plt.xlabel("Data Value")
plt.ylabel("Density")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Map numerical species to actual species names
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Count the number of occurrences of each species
species_count = iris_df['species'].value_counts()

# Create a pie chart for species distribution
plt.figure(figsize=(8, 6))
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.title("Species Distribution in Iris Dataset")
plt.show()
```

# All The Basic Chart Of Matplotlib

## line chart

```python
import matplotlib.pyplot as plt

# Demo data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [100, 120, 90, 110, 130]

# Plotting
plt.plot(months, sales, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales Trend Over Time')

# Displaying the plot
plt.show()
```

## bar chart

```python
# Demo data
products = ['Product A', 'Product B', 'Product C']
sales = [200, 300, 250]

# Plotting
plt.bar(products, sales, color='skyblue')

# Adding labels and title
plt.xlabel('Products')
plt.ylabel('Sales')
plt.title('Sales Comparison Across Products')

# Displaying the plot
plt.show()
```

## Histogram

```python
# Demo data
exam_scores = [85, 90, 70, 95, 80, 75, 88, 92, 78, 85, 90, 82, 87, 89, 93]

# Plotting
plt.hist(exam_scores, bins=5, color='lightgreen', edgecolor='black')

# Adding labels and title
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')

# Displaying the plot
plt.show()
```

## Scatter Plot

```python
# Demo data
study_hours = [3, 4, 5, 6, 7, 8, 9, 10]
exam_scores = [70, 75, 80, 85, 88, 90, 92, 95]

# Plotting
plt.scatter(study_hours, exam_scores, color='orange')

# Adding labels and title
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study Hours vs. Exam Scores')

# Displaying the plot
plt.show()
```

## Pie Chart

```python
# Demo data
regions = ['North', 'South', 'East', 'West']
sales = [30, 20, 25, 35]

# Plotting
plt.pie(sales, labels=regions, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon'])

# Adding title
plt.title('Sales Distribution Across Regions')

# Displaying the plot
plt.show()
```
