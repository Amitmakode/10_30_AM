### - Inner Join

```python
d = {
        'id': ['1', '2', '3', '4', '5','6'],
        'Color': ['RED', 'GREEN', 'YELLOW', 'BLUE', 'PINK','BLACK'],
        'Fruit': ['APPLE', 'BANANA', 'MANGO', 'BERRY', 'MELON','GRAPES']}
d1=pd.DataFrame(d)
d1
```

```python
z = {    'id': ['1', '2', '3', '4', '5','7'],
        'rating': ['A', 'B', 'C', 'A', 'B','A'],
        'COST': [200, 230, 400, 400, 100,450],
        'Fruit': ['APPLE', 'BANANA', 'MANGO', 'BERRY', 'MELON','KIWI'],
        'BUY': ['Y', 'N', 'Y', 'N', 'Y','N']}
d2=pd.DataFrame(z)
d2
```

```python
join = pd.merge(d1, d2, on="Fruit", how='right') ## similary we can  use 'outer/right/left join'

join 
```

```python
join = pd.merge(d1, d2, on="Fruit", how='left') ## similary we can  use 'outer/right/left join'

join 
```

```python
#Selecting few fields from frames
join = pd.merge(d1[["Color","Fruit"]], d2, on="Fruit", how='outer') ## similary we can  use 'inner/right/left join'

join
```

```python
df_merge = pd.merge(d1,d2, on=['Fruit',"id"])

df_merge 
```

```python
#Selecting few fields from frames
join = pd.merge(d1[["Color","Fruit"]], d2, on="Fruit", how='inner') ## similary we can  use 'inner/right/left join'

join
```

```python
#Selecting few fields from frames
join = pd.merge(d1[["Color","Fruit", "id"]], d2, on="id", how='inner') ## similary we can  use 'inner/right/left join'

join
```
