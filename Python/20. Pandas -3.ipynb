## - Inner Join

```python
d = {
        'id': ['1', '2', '3', '4', '5','6'],
        'Color': ['RED', 'GREEN', 'YELLOW', 'BLUE', 'PINK','BLACK'],
        'Fruit': ['APPLE', 'BANANA', 'MANGO', 'BERRY', 'MELON','GRAPES']}
d1=pd.DataFrame(d)
d1
```

```python
z = {    'id': ['1', '2', '3', '4', '5','7'],
        'rating': ['A', 'B', 'C', 'A', 'B','A'],
        'COST': [200, 230, 400, 400, 100,450],
        'Fruit': ['APPLE', 'BANANA', 'MANGO', 'BERRY', 'MELON','KIWI'],
        'BUY': ['Y', 'N', 'Y', 'N', 'Y','N']}
d2=pd.DataFrame(z)
d2
```

```python
join = pd.merge(d1, d2, on="Fruit", how='inner')
```

```python
join 
```

```python
join1 = pd.merge(d1, d2, on="Fruit", how='left')
```

```python
join1
```

```python
join2 = pd.merge(d1,d2, on = "Fruit", how = "right")
```

```python
join2
```

```python
join3 = pd.merge(d1,d2,  how = "cross")
```

```python
join3
```

```python
join4 = pd.merge(d1,d2,  how = "outer")
```

```python
join4
```
