```python
import pandas as pd
import numpy as np
```

```python
# Generating sample data
np.random.seed(0)

# Date ranges
dates = pd.date_range(start='2020-01-01', end='2023-12-31', periods=10000)

# Regions, States, and Cities
regions = np.random.choice(['North', 'South', 'East', 'West'], size=10000)
states = np.random.choice(['StateA', 'StateB', 'StateC', 'StateD'], size=10000)
cities = np.random.choice(["CityV", "CityW", 'CityX', 'CityY', 'CityZ'], size=10000)

# Products and Categories
products = np.random.choice(['ProductA', 'ProductB', 'ProductC', 'ProductD', 'ProductE'], size=10000)
categories = np.random.choice(['Category1', 'Category2', 'Category3'], size=10000)

# Segment, Ship Mode, Shipping Type
segments = np.random.choice(['Consumer', 'Corporate', 'Home Office'], size=10000)
ship_modes = np.random.choice(['Standard Class', 'Second Class', 'First Class'], size=10000)
shipping_types = np.random.choice(['Express', 'Regular', 'Overnight'], size=10000)

# Numeric columns
quantity = np.random.randint(1, 10, size=10000)
profit = np.random.uniform(-25, 100, size=10000).round(2)
sales = np.random.uniform(50, 1000, size=10000).round(2)
discount = np.random.uniform(0, 0.5, size=10000).round(2)
shipping_cost = np.random.uniform(5, 50, size=10000).round(2)
customer_income = np.random.randint(20000, 150000, size=10000)
customer_id = np.random.randint(1000, 5000, size=10000)
```

```python
# Creating the DataFrame
data = {
    'order_date': dates,
    'ship_date': dates + pd.to_timedelta(np.random.randint(1, 10, size=10000), unit='D'),
    'quantity': quantity,
    'profit': profit,
    'sales': sales,
    'region': regions,
    'state': states,
    'city': cities,
    'product': products,
    'product_category': categories,
    'segment': segments,
    'discount': discount,
    'ship_mode': ship_modes,
    'shipping_cost': shipping_cost,
    'shipping_type': shipping_types,
    'customer_income': customer_income,
    'customer_id': customer_id
}

```

## 1: Total Sales by Month

```python
# Group by month and calculate total sales
monthly_sales = df.resample('M', on='order_date')['sales'].sum()
monthly_sales
```

```python
# Plotting
plt.figure(figsize=(8, 5))
monthly_sales.plot(kind='line', marker='x', color='blue', linestyle='-', linewidth=1)
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 2.Profit Distribution by Product Category

```python
# Group by product category and calculate total profit
profit_by_category = df.groupby('product_category')['profit'].sum()

# Plotting
plt.figure(figsize=(8, 4))
profit_by_category.plot(kind='bar', color='yellow')
plt.title('Profit Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Profit')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 3: Shipping Mode Analysis

```python
# Counting orders by ship mode
ship_mode_counts = df['ship_mode'].value_counts()
ship_mode_counts
```

```python
# Plotting as a pie chart
plt.figure(figsize=(5, 5))
plt.pie(ship_mode_counts, labels=ship_mode_counts.index, autopct='%1.2f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Distribution of Orders by Ship Mode')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## 4: Customer Income Distribution

```python
df.columns
```

```python
# Plotting customer income distribution
plt.figure(figsize=(7, 4))
plt.hist(df['customer_income'], bins=10, color='skyblue', edgecolor='black')
plt.title('Customer Income Distribution')
plt.xlabel('Income Level')
plt.ylabel('Number of Customers')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## 5: Relationship between Discount and Profit

```python
# Plotting discount vs. profit
plt.figure(figsize=(8, 6))
plt.scatter(df['discount'], df['profit'], color='green', alpha=0.5)
plt.title('Profit vs. Discount')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 6: Regional Sales Analysis

```python
# Group by region and calculate total sales
sales_by_region = df.groupby('region')['sales'].sum()

# Plotting regional sales
plt.figure(figsize=(3, 3))
sales_by_region.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Plotting regional sales
plt.figure(figsize=(3, 3))
df.loc[:,["region", "sales"]].plot(kind='bar', color='skyblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
